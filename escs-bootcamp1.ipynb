**Prva Monte Karlo Simulacija**

Pretpostaviƒáemo da smo u takmiƒçenju i da nam je dat izbor izmeƒëu tri vrata, gde se iza jednih nalazi automobil kao nagrada u ovom takmiƒçenju. Na nama je da izaberemo vrata. Pretpostavimo da smo odabrali vrata broj 1, a domaƒáin, koji zna ≈°ta se nalazi iza svaki vrata, otvara vrata iza kojih se ne nalazi nagrada (na primer, vrata broj 3). Nakon toga, daje nam izbor: **"Da li ≈æelite da promenite svoj izbor i izaberete vrata broj 2?"**. ≈Ωelimo da otkrijemo ≈°ta bi bilo pametnije za nas - da ostanemo pri inicijalnom izboru ili da ga promenimo.

Ukoliko fiksiramo izbor igraƒça (izbor je pao na Vrata 1) i posmatramo razliƒçite strategije (menjaj/ ne menjaj), 

| Vrata 1 | Vrata 2 |Vrata 3 | Osvajamo, ukoliko ostanemo | Osvajamo, ukoliko promenimo |
| --- | --- | --- | --- | --- | 
| Auto | Koza | Koza | **Auto** | Koza | 
| Koza | Auto | Koza | Koza | **Auto** |
| Koza | Koza | Auto | Koza | **Auto** |

```python
import random

def monte_karlo_simulacije(n):
    
    menjaj = 0
    ne_menjaj = 0
    
    for i in range(n):

        auto = random.randint(1,3)
        izbor1 = random.randint(1,3)
        if (izbor1 == auto): # Izabrali smo vrata iza kojih se nalazi nagrada. Uspeh smo definisali sa 1. 
            izabrana_vrata = 1 # Obja≈°njenje: Ukoliko ostanemo pri svom inicijalnom izboru, osvajamo kola.
            zamenska_vrata = 0 # Ukoliko se odluƒçimo za strategiju "menjaj", izgubiƒáemo ≈°ansu za osvajanje nagrade. 
        else: # Izabrali smo "pogre≈°na" vrata.
            izabrana_vrata = 0 # Ukoliko odluƒçimo da ostanemo pri inicijalnom izboru, gubimo ≈°ansu za osvajanje nagrade.
            zamenska_vrata = 1 # Izabrali smo pogre≈°na vrata (jo≈° to ne znamo). Voditelj otvara druga vrata iza kojih se takoƒëe ne nalazi nagrada.
            # Ukoliko se odluƒçimo da promenimo vrata, osvajamo nagradu.
        menjaj += zamenska_vrata
        ne_menjaj += izabrana_vrata

    return print("Verovatnoƒáa pobede ukoliko se odluƒçimo da promenimo izbor ", menjaj / n,", dok je verovatnoƒáa ukoliko ostanemo pri inicijalnom izboru duplo ni≈æa i jednaka",ne_menjaj/n)

monte_karlo_simulacije(10000)
```
